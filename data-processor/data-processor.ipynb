```python
# import panda libary
import pandas as pd
```

```python
# create a dafaframe 
df = pd.DataFrame()


#then use this dataframe,create a column names "index", assign it with value from 0 to 255

df['index'] = range(0,256)

#then use this dataframe,create a column names "binary", assign it with value from 0 to 255 in binary in string

df['binary'] = df['index'].apply(lambda x: bin(x)[2:].zfill(8))

# create a colume names "is_unit", assign them all to false for now

df['is_unit'] = False

df['unit_index'] = "N/A"

df['rotation'] = 0

df['flip'] = False



```

```python
# create a function used for string bits swapping

def swap_bits(s,rotate_bit_number,flip=False):
    s = s[:]
    rotate_bit_number = 4-rotate_bit_number
    first = s[0:4]
    second = s[4:8]
    #print(first[rotate_bit_number:])
    #print(first[:rotate_bit_number])
    first = first[rotate_bit_number:] + first[:rotate_bit_number]
    second = second[rotate_bit_number:] + second[:rotate_bit_number]
    if flip:
        first_first = first[:2]
        first_second = first[2:]

        second_first = second[:2]
        second_second = second[2:]

        first_first = first_first[::-1]
        first_second = first_second[::-1]
        second_first = second_first[::-1]
        second_second = second_second[::-1]

        first = first_first + first_second
        second = second_first + second_second
        #print(first,second)

    return first + second
```

```python
# write a simple function to test the swap bit function
swap_bits("10110010",1,True)
```

```python
# read lines rotate-1.txt then iterate each line. if the line matches the binary in the dataframe, then assign the is_unit to true

with open('rotate-1.txt') as f:
    for line in f:
        line = line.strip()
        df.loc[df['binary'] == line, 'is_unit'] = True


# read line rotate-2.txt then iterate each line. if the line matches the binary in the dataframe, then assign the is_unit to true.

with open('rotate-2.txt') as f:
    for line in f:
        line = line.strip()
        df.loc[df['binary'] == line, 'is_unit'] = True
        # cacualte the rotate bit number, while rotate bit number is 2
        rotate_bit_number = 2
        temp_str = line[:]
        print(temp_str,line)
        # swap the bits
        temp_bit_str = swap_bits(temp_str+"",0,True)

        print (line,temp_bit_str)
        # if the swap bits matches the binary in the dataframe, then assign the unit_index to the index of the binary
        df.loc[df['binary'] == temp_bit_str, 'unit_index'] = line
        # flip equals to true
        df.loc[df['binary'] == temp_bit_str, 'flip'] = True


# write the dataframe to a csv file
df.to_csv('output.csv',index=False)




```
